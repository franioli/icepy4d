```python
import os

from pathlib import Path
from tqdm import tqdm
import gc
import linecache
import tracemalloc
from time import sleep
import os
from collections import Counter
import psutil

# CloudComPy
# os.environ["_CCTRACE_"]="ON"          # only if you want debug traces from C++
import cloudComPy as cc               # import the CloudComPy module

pcd_dir = Path("../res/point_clouds")

i = 0
tstep = 3

# Read point cloud list
assert pcd_dir.is_dir(), "Directory does not exists."
pcd_lst = sorted(pcd_dir.glob("dense_*.ply"))

i = 0
tstep = 3

# Big Loop

result = []

verbose = 1

# tracemalloc.start()
process = psutil.Process(os.getpid())

for i in range(3): # tqdm(range(len(pcd_lst) - tstep)):
    
    # Load point cloud i
    pcd0 = cc.loadPointCloud(str(pcd_lst[i]))

    # Load point cloud i+tstep
    pcd1 = cc.loadPointCloud(str(pcd_lst[i+tstep]))

    # Compute variation by DOD: pcd(i) - pcd(i+step)
    if verbose:
        print(f"Computing volume variation between point clouds {str(pcd_lst[i])} and {str(pcd_lst[i+tstep])}")
    result = cc.ReportInfoVol()
    isOk = cc.ComputeVolume25D(result, ground=pcd0, ceil=pcd1,
                               vertDim=0, gridStep=0.20, groundHeight=0, ceilHeight=0) 
    if not isOk:
        raise RuntimeError(f"Unable to compute volume variation between point clouds {str(pcd_lst[i])} and {str(pcd_lst[i+tstep])}")
    
    if verbose:
        print(f"""Volume variation result:
        Volume: {result.volume:.2f} m3
        Added volume: {result.addedVolume:.2f} m3
        Removed volume: {result.removedVolume:.2f} m3
        Surface: {result.surface:.2f} m2
        Maching Percent {result.matchingPercent:.1f}%
        Average Neighbora per cell: {result.averageNeighborsPerCell:.1f}
        """)
    
    # snapshot = tracemalloc.take_snapshot()
    # display_top(snapshot)    
    
    # Remove variables and free memory     
    del pcd0, pcd1
    gc.collect()

    # g = globals()
    # print(g)
    
    print(f"Memory usage: {psutil.Process(os.getpid()).memory_info().rss / 1024 ** 2}")
        
    # snapshot = tracemalloc.take_snapshot()
    # display_top(snapshot)    
```
